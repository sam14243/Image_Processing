Generate the edges of an image by applying the following edge detection operators: 

1.Roberts 

2.Sobel 

3.Prewitt

4.Laplacian 

5.Canny

```python
from scipy import ndimage 
import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage
```

```python
img = cv2.imread("Butterfly.jpg")
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img, cmap='gray')
```

```python
edge = skimage.filters.roberts(img)
plt.imshow(edge, cmap='gray')
```

```python
edge = skimage.filters.sobel(img)
plt.imshow(edge, cmap='gray')
```

```python
edge = skimage.filters.prewitt(img)
plt.imshow(edge, cmap='gray')
```

```python
edges = cv2.Laplacian(img, -1, ksize=3, scale=1, delta=0,borderType=cv2.BORDER_DEFAULT)
plt.imshow(edges, cmap='gray')
```

```python
edges = cv2.Canny(img,100,200)
plt.imshow(edges, cmap='gray')
```
